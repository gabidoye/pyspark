```python
import findspark
findspark.init("/Users/DOU2274/spark/spark-3.1.1-bin-hadoop2.7")

from pyspark.sql.types import *
from pyspark.sql import SparkSession
import warnings
warnings.filterwarnings('ignore')
```

```python
spark = SparkSession.builder.appName("temprature calculation").getOrCreate()
```

```python
schema = StructType([StructField("celsius", ArrayType(IntegerType()))])
t_list = [[35, 36, 32, 30, 40, 42, 38]], [[31, 32, 34, 55, 56]] 
```

```python
t_c = spark.createDataFrame(t_list, schema) 

t_c.createOrReplaceTempView("tC")
```

```python
t_c.show()
```

```python
spark.sql("""SELECT celsius,transform(celsius, t -> ((t * 9) div 5) + 32) as fahrenheit FROM tC""").show()
```

```python
#Filter()
spark.sql("""
SELECT celsius,
filter(celsius, t -> t > 38) as high FROM tC
""").show()
```

```python
#Exist()
#// Is there a temperature of 38C in the array of temperatures 
spark.sql("""
SELECT celsius,
exists(celsius, t -> t = 38) as threshold FROM tC
""").show()
```
